```python
class calc():
    
    def __init__(self, x, y):
        self.num1 = x
        self.num2 = y
    
    def add(self):
        return self.num1 + self.num2
    
    
    def subtract(self):
        return self.num1 - self.num2
    
    
    def multiply(self):
        return self.num1 * self.num2
    
    
    def divide(self):
        return self.num1 / self.num2


while True:
    print("Select operation.")
    print("1.Add")
    print("2.Subtract")
    print("3.Multiply")
    print("4.Divide")
    
    choice = input("Enter choice(1/2/3/4): ")
    
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))
    op_obj = calc(num1, num2)
    
    if choice in ('1', '2', '3', '4'):
        if choice == '1':
            print(num1, "+", num2, "=", op_obj.add())

        elif choice == '2':
            print(num1, "-", num2, "=", op_obj.subtract())

        elif choice == '3':
            print(num1, "*", num2, "=", op_obj.multiply())

        elif choice == '4':
            print(num1, "/", num2, "=", op_obj.divide())
        
        
        next_calculation = input("Let's do next calculation? (yes/no): ")
        if next_calculation == "no":
          break
    
    else:
        print("Invalid Input")
```
